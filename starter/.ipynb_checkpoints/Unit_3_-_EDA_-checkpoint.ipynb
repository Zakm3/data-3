#### Fill in blanks with a brief description of each plot type:

  
- Histogram
> A histogram is a chart that groups numeric data into bins, displaying the bins as segmented columns.
- Density Plot / KDE
> KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable.
- Bar Chart
> Bar charts are one of the many techniques used to present data in a visual form so that the reader may readily recognize patterns or trends. Bar charts usually present categorical variables, discrete variables or continuous variables grouped in class intervals.
- Line Plot
> a line plot is a graph that uses lines to connect individual data points 
- Univariate scatter plot
> Plots different observations/values of the same variable corresponding to the index/observation number.
- Violin Plot
> A violin plot depicts distributions of numeric data for one or more groups using density curves.
- Box Plot
> A box plot is a graphical rendition of statistical data based on the minimum, first quartile, median, third quartile, and maximum. The term "box plot" comes from the fact that the graph looks like a rectangle with lines extending from the top and bottom.
- Pie Chart
> A pie chart  is a circular statistical graphic, which is divided into slices to illustrate numerical proportion 
- Rug Plot
> A Rug Plot would be another way to display a distribution as an adjunct to your primary plot.  A Rug Plot is simply a projection of raw data points onto a particular axis, without involving the binning process of a histogram.

### Step 5 - Create Plots for Each Variable in Dataset

```python
# import seaborn
import seaborn as sns


# set standard plot type as seaborn plot
sns.set()


# import matplotlib
import matplotlib.pyplot as plt


# write command for plots to appear in jupyter cells automatically

%matplotlib inline
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

### Plot for 'age' column

```python
# create plot for age column
# sets figure size
plt.figure(figsize=(14,7))

# establishes figure with data and plot type
ax = sns.histplot(data=df, x='age', palette="magma",  bins=8)

# sets x and y labels
plt.xlabel("Age", fontsize=20)
plt.ylabel("Number of Managers", fontsize=14)

# sets title
plt.title("Manager Ages in Survey", fontsize=15);
```

##### Insights:

We don't have as many young managers as we do old ones.

### Plot for 'industry' column

```python
# create plot for industry column
plt.figure(figsize=(10,10))

data = df['industry'].value_counts()

labels = ["Healthcare", "Technology", "Finance", "Business"]

colors = sns.color_palette('husl')

plt.title("Percent of Managers in each Industry", fontsize=16)

ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%');
```

##### Insights:

An equal amount of managers in all gathered industry fields.

### Plot for 'salary' column

```python
# create plot for salary column
plt.figure(figsize=(14,7))

sns.violinplot(x=df['salary'], color = 'skyblue')

plt.xlabel("Salary", fontsize=14)
plt.title("Density of Manager Salaries");
```

##### Insights:

Most of the manager salaries are in the 100,000 range

### Plot for 'industry_experience' column

```python
# create plot for industry experience column
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='industry_experience',palette="rocket")

plt.xticks(rotation=90)
plt.xlabel("Industry experience", fontsize=13)
plt.ylabel("Number of Managers", fontsize=13)
plt.title("Industry Experience of Managers", fontsize=15);
```

##### Insights:

Most of the Managers that took the survey have industry experince of 15-25 yesrs.

### Plot for 'education' column

```python
plt.figure(figsize=(10,10))

data = df['education'].value_counts()

labels = ["bachelor's degree", "master's degree", "some college", "high school"]

colors = sns.color_palette('crest')

plt.title("Manager's Education Levels", fontsize=16)

ax = plt.pie(data, colors = colors, labels = labels, autopct = '%.0f%%');
```

##### Insights:

Most managers have a Bachelors or Master's degree

### Plot for 'gender' column

```python
# create plot for gender column
plt.figure(figsize=(12,6))

ax = sns.countplot(data=df, x='gender',palette="icefire")

plt.xticks(rotation=90)
plt.xlabel("Gender", fontsize=13)
plt.ylabel("Number of Each Gender in Industry", fontsize=13)
plt.title("Gender of managers in Survey", fontsize=15);
```

##### Insights:

An equal amount of Male and Female managers took the survey.

### Step 1 - Review Summary Statistics

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

- mean
> The mean is the average of a data set.
- median
> the median is the value separating the higher half from the lower half of a data sample
- mode
> The mode is the value that appears most often in a set of data 
- central tendency
> A measure of central tendency is a single value that attempts to describe a set of data by identifying the central position within that set of data
- distribution
> A statistical distribution, or probability distribution, describes how values are distributed for a field. In other words, the statistical distribution shows which values are common and uncommon
- variance
> The term variance refers to a statistical measurement of the spread between numbers in a data set
- standard deviation

> Standard deviation is a statistic that measures the dispersion of a dataset relative to its mean and is calculated as the square root of the variance. The standard deviation is calculated as the square root of variance by determining each data point's deviation relative to the mean.

### Step 3 - Use Pearsonr & Spearmanr To Answer Questions About Correlation and P-Values

<h3 style='color:red;'> First, load 'salary_numeric.csv' - where education has been re-expressed as numeric variables.</h3>

```python
# load salary_numeric.csv into dataframe where education levels have been re-expressed as numeric
df = pd.read_csv('salary_numeric.csv')
```

```python
################################### TO DO #########################################
####### FOLLOW PROJECT INSTRUCTIONS TO COMPLETE THIS STEP IN THE CELLS BELOW ######
```

```python
# import pearson and spearmanr methods from scipy.stats library
from scipy.stats import pearsonr, spearmanr
```

### What is the correlation between salary and education?

#### Your Hypothesis = 
> higher education, higher salary

```python
# calculate correlation and pvalue of salary and education with spearmanr method
spearmanr(df['salary'], df['education'])
```

A positve correlation between education level and salary. 

#### Are these results statistically significant? How do you know?

Yes, because the pvalue is 0. Anything lower than .05 is considered significant.

### What is the correlation between salary and age?

#### Your Hypothesis 
> I think there is a positive correlation between salary and age

```python
# calculate correlation and pvalue of salary and age with pearsonr method
pearsonr(df['salary'], df['age'])
```

There is a positive correlation between salary and age.

#### Are these results statistically significant? How do you know? 

Yes, because the pvalue is 0. Anything lower than .05 is considered significant.

### What is the correlation between salary and industry experience? 

#### Your Hypothesis  
> There is a positive correlation between salary and industry experience

```python
# calculate correlation and pvalue of salary and industry experience with pearsonr method
spearmanr(df['salary'], df['industry_experience'])
```

There is a positive correlation between salary and industry experience

#### Are these results statistically significant? How do you know?

Yes, because the pvalue is 0. Anything lower than .05 is considered significant.

##### Insights:

#### Crosstab of average salary grouped by education and industry

```python
# use pd.crosstab() to calculate mean salary in each industry grouped by education level
pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

#### Heatmap of crosstab of average salary grouped by education and industry

```python
# save above result to new variable
heat_crosstab = pd.crosstab(index=df['industry'], columns=df['education'], values=df['salary'], aggfunc=np.mean).round(0)
```

```python
# create heatmap with annotated mean salary in each industry and education level
plt.figure(figsize=(16,10))

ax = sns.heatmap(heat_crosstab, annot=True, linewidths=.5, fmt='g', xticklabels=['High School', 'Some College', "Bachelor's Degree", "Master's Degree"])

plt.xlabel("Education Level", fontsize=14)
plt.ylabel("Industry", fontsize=14)

plt.title("Manager Salaries Across Industries by Education Level", fontsize=18);
```

##### Insights:

With a darker the color it's a more distant relationship and a lighter color means they are closely related
